```python
def fourier_transform(img_path):
    # Load the image in grayscale
    img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
    
    # Check if image was loaded successfully
    if img is None:
        print(f"Error: Could not load image from {img_path}")
        return
    
    # Perform the 2D Fourier Transform
    f = np.fft.fft2(img)
    fshift = np.fft.fftshift(f)  # Shift zero frequency components to the center

    # Get the magnitude spectrum
    magnitude_spectrum = np.abs(fshift)

    # Display original image and its magnitude spectrum
    plt.figure(figsize=(12, 6))

    plt.subplot(1, 2, 1)
    plt.imshow(img, cmap='gray')
    plt.title('Original Image')
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.imshow(np.log(1 + magnitude_spectrum), cmap='gray')
    plt.title('Magnitude Spectrum (Fourier)')
    plt.axis('off')

    plt.show()
```

```python
fourier_transform('cv_lab_model3.jpeg')
```
