```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load an image
img = cv2.imread('lab-3.jpeg')  # Replace with your image path

# Check if the image was loaded successfully
if img is None:
    print("Error: Image not found!")
    exit()

# 1. Drawing a Line
# Parameters: start point, end point, color (BGR), thickness
start_point = (50, 50)
end_point = (400, 50)
line_color = (0, 0, 255)  # Red color in BGR
line_thickness = 3
cv2.line(img, start_point, end_point, line_color, line_thickness)

# 2. Drawing a Rectangle
# Parameters: top-left corner, bottom-right corner, color (BGR), thickness
top_left = (50, 100)
bottom_right = (400, 200)
rectangle_color = (0, 255, 0)  # Green color in BGR
rectangle_thickness = 2
cv2.rectangle(img, top_left, bottom_right, rectangle_color, rectangle_thickness)

# 3. Drawing a Circle
# Parameters: center, radius, color (BGR), thickness
center = (250, 350)
radius = 50
circle_color = (255, 0, 0)  # Blue color in BGR
circle_thickness = -1  # Fill the circle
cv2.circle(img, center, radius, circle_color, circle_thickness)

# 4. Drawing an Ellipse
# Parameters: center, axes lengths, angle of rotation, start angle, end angle, color (BGR), thickness
center = (250, 500)
axes = (100, 50)  # Lengths of the axes
angle = 0
start_angle = 0
end_angle = 180  # Half ellipse
ellipse_color = (255, 255, 0)  # Yellow color in BGR
ellipse_thickness = 2
cv2.ellipse(img, center, axes, angle, start_angle, end_angle, ellipse_color, ellipse_thickness)

# 5. Adding Text
# Parameters: position, text, font, font size, color (BGR), thickness
text = "OpenCV Annotations"
position = (50, 600)
font = cv2.FONT_HERSHEY_SIMPLEX
font_scale = 1
text_color = (255, 255, 255)  # White color in BGR
text_thickness = 2
cv2.putText(img, text, position, font, font_scale, text_color, text_thickness)

# Convert the image to RGB for displaying with Matplotlib
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Display the annotated image
plt.imshow(img_rgb)
plt.title('Image with Annotations')
plt.axis('off')
plt.show()
```
