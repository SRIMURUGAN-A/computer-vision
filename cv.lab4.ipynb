```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load an image
img = cv2.imread('cv_lab_model3.jpeg')  # Replace with the path to your image

# Check if the image was loaded successfully
if img is None:
    print("Error: Image not found!")
    exit()

# ================================
# 1. Color Space Conversion
# ================================
# Convert RGB to HSV
hsv_img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

# Convert RGB to Lab
lab_img = cv2.cvtColor(img, cv2.COLOR_BGR2LAB)

# Convert RGB to Grayscale
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Display the images
plt.figure(figsize=(12, 6))

plt.subplot(1, 3, 1)
plt.imshow(cv2.cvtColor(hsv_img, cv2.COLOR_BGR2RGB))
plt.title('HSV Image')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(cv2.cvtColor(lab_img, cv2.COLOR_BGR2RGB))
plt.title('Lab Image')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(gray_img, cmap='gray')
plt.title('Grayscale Image')
plt.axis('off')

plt.show()

# ================================
# 2. Histogram Equalization
# ================================
# Apply histogram equalization to the grayscale image
equalized_img = cv2.equalizeHist(gray_img)

# Display original and equalized images
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.imshow(gray_img, cmap='gray')
plt.title('Original Grayscale')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(equalized_img, cmap='gray')
plt.title('Equalized Grayscale')
plt.axis('off')

plt.show()

# ================================
# 3. Convolution (Sharpening)
# ================================
# Define a kernel for sharpening
sharpening_kernel = np.array([[0, -1, 0],
                               [-1, 5, -1],
                               [0, -1, 0]])

# Apply convolution (sharpening) to the image
sharpened_img = cv2.filter2D(img, -1, sharpening_kernel)

# Display original and sharpened images
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(sharpened_img, cv2.COLOR_BGR2RGB))
plt.title('Sharpened Image')
plt.axis('off')

plt.show()

# ================================
# 4. Image Smoothing (Gaussian Blur and Median Filter)
# ================================
# Apply Gaussian blur
gaussian_blurred_img = cv2.GaussianBlur(img, (5, 5), 0)

# Apply Median blur
median_blurred_img = cv2.medianBlur(img, 5)

# Display original, Gaussian blurred, and Median blurred images
plt.figure(figsize=(12, 6))

plt.subplot(1, 3, 1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(cv2.cvtColor(gaussian_blurred_img, cv2.COLOR_BGR2RGB))
plt.title('Gaussian Blurred Image')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(cv2.cvtColor(median_blurred_img, cv2.COLOR_BGR2RGB))
plt.title('Median Blurred Image')
plt.axis('off')

plt.show()

# ================================
# 5. Gradients (Sobel Operator)
# ================================
# Compute Sobel gradients in x and y directions
sobel_x = cv2.Sobel(gray_img, cv2.CV_64F, 1, 0, ksize=3)
sobel_y = cv2.Sobel(gray_img, cv2.CV_64F, 0, 1, ksize=3)

# Compute the gradient magnitude
gradient_magnitude = cv2.magnitude(sobel_x, sobel_y)

# Display gradient magnitude
plt.imshow(gradient_magnitude, cmap='gray')
plt.title('Gradient Magnitude')
plt.axis('off')
plt.show()

# ================================
# 6. Edge Detection (Canny Edge Detection)
# ================================
# Apply Canny edge detection
edges = cv2.Canny(gray_img, 100, 200)

# Display the edge-detected image
plt.imshow(edges, cmap='gray')
plt.title('Canny Edge Detection')
plt.axis('off')
plt.show()
```
